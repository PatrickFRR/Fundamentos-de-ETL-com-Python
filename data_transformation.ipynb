```python
import pandas as pd
import pandera as pa
```

```python
data = pd.read_csv("ocorrencia.csv", sep=";", parse_dates=['ocorrencia_dia'], dayfirst=True)
data = data.drop(['codigo_ocorrencia1', 'codigo_ocorrencia3','codigo_ocorrencia4', 'ocorrencia_latitude', 'ocorrencia_longitude', 
          'ocorrencia_pais', 'investigacao_aeronave_liberada','investigacao_status', 'divulgacao_relatorio_numero',
          'divulgacao_relatorio_publicado', 'divulgacao_dia_publicacao','total_aeronaves_envolvidas', 
          'ocorrencia_saida_pista'], axis=1)
data.head()
```

```python
data.replace(['**','###!','####','****','*****','NULL'], pd.NA, inplace=True)
```

```python
data
```

```python
schema = pa.DataFrameSchema(
    columns = {
        "codigo_ocorrencia": pa.Column(pa.Int),
        "codigo_ocorrencia2": pa.Column(pa.Int),
        "ocorrencia_classificacao": pa.Column(pa.String),
        "ocorrencia_cidade": pa.Column(pa.String),
        "ocorrencia_uf": pa.Column(pa.String, pa.Check.str_length(2,2), nullable=True),
        "ocorrencia_aerodromo": pa.Column(pa.String, nullable=True),
        "ocorrencia_dia": pa.Column(pa.DateTime),
        "ocorrencia_hora": pa.Column(pa.String, pa.Check.str_matches(r'^([0-1]?[0-9]|[2][0-3]):([0-5][0-9])(:[0-5][0-9])?$'), nullable=True),
        "total_recomendacoes": pa.Column(pa.Int)   
    }
)
```

```python
schema.validate(data)
```

```python
# ocorrências com mais de 10 recomendações
filtro = data.total_recomendacoes > 10
data.loc[filtro]
```

```python
# ocorrências com mais de 10 recomendações
filtro = data.total_recomendacoes > 10
data.loc[filtro, ['ocorrencia_cidade', 'total_recomendacoes']]
```

```python
# ocorrências cuja classificação == INCIDENTE GRAVE	
filtro = data.ocorrencia_classificacao == 'INCIDENTE GRAVE'
data.loc[filtro]
```

```python
# ocorrências cuja classificação == INCIDENTE GRAVE e o estado == SP
filtro1 = data.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = data.ocorrencia_uf == 'SP'
data.loc[filtro1 & filtro2]
```

```python
# ocorrências cuja classificação == INCIDENTE GRAVE ou o estado == SP
filtro1 = data.ocorrencia_classificacao == 'INCIDENTE GRAVE'
filtro2 = data.ocorrencia_uf == 'SP'
data.loc[filtro1 | filtro2]
```

```python
# ocorrências cuja (classificação == INCIDENTE GRAVE ou classificação == INCIDENTE) e o estado == SP
filtro1 = data.ocorrencia_classificacao.isin(['INCIDENTE GRAVE', 'INCIDENTE'])
filtro2 = data.ocorrencia_uf == 'SP'
data.loc[filtro1 & filtro2]
```

```python
# ocorrências cuja cidade comecem com a letra C
filtro = data.ocorrencia_cidade.str[0] == 'C'
data.loc[filtro]
```

```python
# ocorrências cuja cidade terminam com a letra A
filtro = data.ocorrencia_cidade.str[-1] == 'A'
data.loc[filtro]
```

```python
# ocorrências cuja cidade terminam com os caracteres MA
filtro = data.ocorrencia_cidade.str[-2:] == 'MA'
data.loc[filtro]
```

```python
# ocorrências cuja cidade contém (em qualquer parte do conteúdo) os caracteres MA ou AL
filtro = data.ocorrencia_cidade.str.contains('MA|AL')
data.loc[filtro]
```

```python
# ocorrências do ano de 2015
filtro = data.ocorrencia_dia.dt.year == 2015
data.loc[filtro]
```

```python
#ocorrências do ano de 2015 e mês 12 e dias entre 3 e 8
filtro_ano = data.ocorrencia_dia.dt.year == 2015
filtro_mes = data.ocorrencia_dia.dt.month == 12
filtro_dia_inicio = data.ocorrencia_dia.dt.day > 2 
filtro_dia_fim = data.ocorrencia_dia.dt.day < 9
data.loc[filtro_ano & filtro_mes & filtro_dia_inicio & filtro_dia_fim]
```
